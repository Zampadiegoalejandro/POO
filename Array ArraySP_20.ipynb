```python
class ProcesadorArray():
  def __init__(self):
    r=0

  def obtenerArray(self):
    a = [
        ['5', '6', '9', 'Z'],
        ['3', 'R', '89', '44'],
        ['45', 'A4', '7', '56'],
        ['44', '45', 'X', '34'],
        ['Y', '5', '6', '7']
    ]
    return(a)

  def extraer_letras(self, array):
    letras = []
    for fila in array:
        for elemento in fila:
            if any(c.isalpha() for c in str(elemento)):
                letras.append(elemento)
    return(letras)

  def extraer_numeros(self, array):
    numeros = []
    for fila in array:
        for elemento in fila:
            if str(elemento).isdigit():
                numeros.append(int(elemento))
    return(numeros)

  def encontrar_mayor_menor(self, numeros):
    if numeros:
        return max(numeros), min(numeros)
    return None, None

  def imprime(self, array, letras, mayor, menor):
    print("\nArray original:")
    print(array)
    print("\nLetras encontradas en el array:")
    print(letras)
    print("\nNúmero Mayor:", mayor)
    print("Número Menor:", menor)


procesador = ProcesadorArray()
array = procesador.obtenerArray()
letras = procesador.extraer_letras(array)
numeros = procesador.extraer_numeros(array)
mayor, menor = procesador.encontrar_mayor_menor(numeros)
procesador.imprime(array, letras, mayor, menor)
```

```python
class ProcesadorArraySP():
  def __init__(self):
    self.array = [
        ['5', '6', '9', 'Z'],
        ['3', 'R', '89', '44'],
        ['45', 'A4', '7', '56'],
        ['44', '45', 'X', '34'],
        ['Y', '5', '6', '7']
    ]
    self.letras = []
    self.numeros = []
    self.mayor = 0
    self.menor = 0

  def Menu(self):
    print("\nElije una Opcion")
    print("1. Extraer letras")
    print("2. Extraer números")
    print("3. Encontrar mayor y menor")
    print("4. Mostrar todos los resultados")
    print("5. Salir")
    op = int(input("Opcion: "))
    return(op)

  def extraer_letras(self):
      self.letras = []
      for fila in self.array:
          for elemento in fila:
            elemento_str = str(elemento)
            for c in elemento_str:
              if ('a' <= c <= 'z') or ('A' <= c <= 'Z'):
                self.letras.append(elemento)
                break

  def extraer_numeros(self):
    self.numeros = []
    for fila in self.array:
      for elemento in fila:
        try:
          numero = int(elemento)
          self.numeros.append(numero)
        except ValueError:
            pass


  def encontrar_mayor_menor(self):
    if self.numeros:
        self.mayor = max(self.numeros)
        self.menor = min(self.numeros)
    else:
        self.mayor = None
        self.menor = None

  def mostrar_resultados(self):
    print("\nArray original:")
    for fila in self.array:
        print(fila)

    print("\nLetras encontradas en el array:")
    print(self.letras)

    print("\nNúmeros encontrados en el array:")
    print(self.numeros)

    print("\nNúmero Mayor:", self.mayor)
    print("Número Menor:", self.menor)

procesador = ProcesadorArraySP()

while True:
  try:
    opcion = procesador.Menu()

    if opcion == 1:
      procesador.extraer_letras()
      print("\nLetras extraídas:")
      print(procesador.letras)

    elif opcion == 2:
      procesador.extraer_numeros()
      print("\nNúmeros extraídos:")
      print(procesador.numeros)

    elif opcion == 3:
      if not procesador.numeros:
        procesador.extraer_numeros()
      procesador.encontrar_mayor_menor()
      print("\nNúmero Mayor:", procesador.mayor)
      print("Número Menor:", procesador.menor)

    elif opcion == 4:
      procesador.extraer_letras()
      procesador.extraer_numeros()
      procesador.encontrar_mayor_menor()
      procesador.mostrar_resultados()

    elif opcion == 5:
      print("\nPrograma finalizado")
      break

    else:
      print("\nOpción inválida, intente de nuevo")

  except ValueError:
    print("\nError: Ingrese un número válido")
```
